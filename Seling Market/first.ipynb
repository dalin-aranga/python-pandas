```python
import pandas as pd
import os

list=[]
for file in os.listdir('/home/dalin/jupyter/Sales_Data'):
    list.append(file)
```

```python
df = pd.DataFrame()
for file in list:
    df1 = pd.read_csv('/home/dalin/jupyter/Sales_Data/'+file)
    df=pd.concat([df,df1])
df
```

```python
df.tail()
```

```python
df
```

### best month in sales

```python
df['Month'] = df["Order Date"].str[0:2]
df
```

```python
df = df[df['Order Date'].str[0:2]!='Or']
df = df.dropna()
df
```

```python
df['Month'] = df['Month'].astype('int32')
df
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

```python
month = df.groupby('Month').sum()
month
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13,1)
plt.bar(months,month['Sales'])
plt.xticks(months)
plt.xlabel('Month')
plt.ylabel('Sales each month $ bilion')
plt.show()
```

### best sale city

```python
df
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1])
df
```

```python
df['City_leter'] = df['Purchase Address'].apply(lambda x: x.split(',')[2])
df
```

```python
df['City'] = df['City']+ '(' +df['City_leter'].str[1:3] +')'
df = df.drop('City_leter',axis=1)
df
```

```python
df
```

```python
city = df.groupby('City').sum()
city
```

```python
cites = [city for city, df in df.groupby('City')]
plt.plot(cites,city['Sales'])
plt.xticks(cites,rotation=90)
plt.xlabel('Name Of The city')
plt.ylabel(' Sales amount Bilion($)')
plt.show()
```

```python
cites = [city for city, df in df.groupby('City')]
plt.bar(cites,city['Sales'])
plt.xticks(cites,rotation='vertical')
plt.xlabel('Name Of The city')
plt.ylabel(' Sales amount Bilion($)')
```
